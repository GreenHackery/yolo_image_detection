```
from ultralytics import YOLO
import os
import yaml

# Base directory for your dataset
DATASET_DIR = "/home/daniel/projects/foodlikeair/yolo_image_detection/strawberry_dataset"  # change to your path
```

```
data_yaml = {
    'path': os.path.abspath(DATASET_DIR),  # root dir
    'train': 'images/train',
    'val': 'images/val',
    'names': {
      0: 'dead',
      1: 'overripe',
      2: 'ripe',
      3: 'strawberry',
      4: 'stem',
      5: 'stem_leaves',
      6: 'underripe'
    }
}

yaml_path = os.path.join(DATASET_DIR, "data.yaml")
with open(yaml_path, 'w') as f:
    yaml.dump(data_yaml, f, default_flow_style=False)

print(f"Config written to {yaml_path}")
print(yaml.dump(data_yaml))
```

```
# yolov8n.pt = nano (fastest, least accurate)
# yolov8s.pt = small
# yolov8m.pt = medium (good balance)
model = YOLO("yolov8m.pt")
```

```
results = model.train(
    data=yaml_path,
    epochs=100,
    imgsz=640,
    batch=16,
    patience=20,         # early stopping if no improvement
    project="strawberry_runs",
    name="strawberry_detector",
    exist_ok=True,
    device='cpu',             # use 0 for GPU, 'cpu' for CPU-only
)
```
